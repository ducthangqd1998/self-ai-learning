### Optimization Algorithms (Thuật toán tối ưu)

**Thuật toán tối ưu** là các thuật toán toán học được sử dụng nhằm tìm ra những giá trị (weight + bias) tốt nhất có thể , nhằm nâng cao độ chính xác các mô hình học máy.

Việc tìm các giá trị này là cực kì khó khăn và không thể tìm một cách ngẫu nhiên hay là xác định được trong lần đầu tiên. Chúng ta cần những thuật toán tìm ra và tối ưu (cập nhật) từ từ các giá trị đó.

**Gradient Descent** là một thuật toán tối ưu lặp (iterative optimization algorithm) được sử dụng trong các bài toán học máy và học sâu (thường là các bài toán tối ưu lồi — Convex Optimization) với mục tiêu là tìm một tập các biến nội tại (internal parameters) cho việc tối ưu mô hình.

**Trong đó**
* Gradient: là tỷ lệ độ nghiêng của đường dốc (rate of inclination or declination of a slope). Về mặt toán học, Gradient của một hàm số là đạo hàm của hàm số đó tương ứng với mỗi biến của hàm. Đối với hàm số đơn biến, chúng ta sử dụng khái niệm Derivative thay cho Gradient.
* Descent: là từ viết tắt của descending, nghĩa là giảm dần.

### Gradient cho hàm 1 biến 

Giả sử cho một hàm $y = f(x)$, ta có $x_t$ là giá trị tìm được sau vòng lặp thứ $t$. Và ta cần thuật toán đưa $x_t$ càng về gần $x'$ ($x'$ là giá trị mà đạo hàm $f$ bằng 0)

Nếu đạo hàm của hàm số tại $x_t$ đồng biến, thì $x_t > x'$, thì giá trị $x_{t+1}$ phải giảm, tức sẽ di chuyển về phía âm, ngược dấu với đạo hàm. Nếu đạo hàm của hàm số tại $x_t$ nghịch biến, thì $x_t < x'$, thì giá trị $x_{t+1}$ phải tăng, tức sẽ di chuyển về phía dương, ngược dấu với đạo hàm. Nếu $x_t$ càng xa $x'$ thì $f'(x_t)$ càng lớn và ngược lại. Như vậy giá trị $x_{t+1} = x_t - \Delta f'(x_t)$ 

**Công thức cập nhật trọng số cho GD là:**
$$x_{t+1} = x_t - \Delta f'(x_t)$$

### Gradient cho hàm nhiều biến
Giả sử ta có tập $\theta$ là tập các trọng số cần tối ưu cho hàm $f(\theta)$. Thì gradient của hàm số đó tại một điểm $\theta$ bất kỳ được lý hiệu là $\nabla_\theta f(\theta)$. Như vậy với hàm nhiều biến cũng bắt đầu bằng một điểm dự đoán $\theta_0$, sau đó sử dụng quy tắc cập nhật.
$$\theta_{t+1} = \theta_t - \Delta \nabla_\theta f(\theta_t)$$
Cách viết đơn giản hơn là:
$$\theta \leftarrow \theta - \Delta \nabla_\theta f(\theta)$$



**Tối ưu hàm mất mát** là việc tìm các điểm cực trị, mà ở đó hàm mất mát có thể
là cực đại (maximum) hoặc cực tiểu (minimum). Nếu hàm mất mát mà không
phải là hàm lồi thì sẽ có các giá trị cực đại địa phương (local maximum) và cực
tiểu địa phương (local minimum) bên cạnh các cực đại toàn cục (global
maximum) và cực tiểu toàn cục (global minimum). Mục tiêu của hàm mất mát là
tìm ra được các cực tiểu toàn cục.

+ Learning rate càng lớn thì tốc độ tiến xuống dốc (descent) càng nhanh,
nhưng đồng thời càng khó đi sâu (deep) xuống dốc.
