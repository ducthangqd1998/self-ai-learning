### Perceptron

**Perceptron** là một thuật toán học máy cơ bản, là nền tảng quan trọng cho Neural Networks và sau này là học sâulà một thuật toán phân loại nhị phân (binary classifier), là một kiểu của phân loại tuyến tính (linear classifier)

**Linear Classifier**: Là phân loại các dữ liệu huấn luyện thành các category tương ứng dựa vào các hàm dự đoán tuyến tính kết hợp với một tập trọng số  với vector đặc trưng. Hay nói cách khác, thuật toán mà tạo ra các đường 
biên là các đường thẳng phân loại các điểm dữ liệu là Linear Classifier.

**Binary Classifier**: 1 đường thẳng và 2 lớp.


**Trong đó**

* **Input**: Là tập hợp các đặc trưng đầu vào của mạng. Kí hiệu ở đây là $[x_1, x_2, ..., x_n]$. Giá trị **1** ở đây là hằng số
* **Weights**: Là các trọng số huấn luyện, được khởi tạo ngẫu nhiên và cập nhật sau mỗi lần huấn luyện. Kí hiệu $[w_1, w_2, ..., w_n]$. Giá trị đặc biệt $w_0$ được gọi là **bias**, nó cho phép bộ phân loại (classifier) dịch sang trái hoặc phải (boundary). Trong toán học thì neural bias cho phép bộ phân loại có thể di chuyển boundary của nó, di chuyển mọi điểm một khoảng cách không đổi và theo một hướng xác định. Về mặt huấn luyện thì nó giúp huấn luyện mô hình nhanh hơn và chính xác hơn. 
* **Weighted Summation**: Là phép tổng của các giá trị của phép tính tích giữa đầu vào thứ $i$ với trọng số tương ứng của nó, ví dụ $x_i w_i$. Phép tính tương ứng là $\sum_{i=1}^n x_i w_i + bias$
* **Step/Activation/Transfer Function**: Là hàm kích hoạt khiến mô hình trở thành một hàm phi tuyến.

### Neural Network 

**Neural Network**: là mô hình học máy được liên kết giữa nhiều perceptron. Neural network còn gọi là multi-layer Perceptron (MLP)
<td> <img src="https://miro.medium.com/max/1100/1*YgJ6SYO7byjfCmt5uV0PmA.png" alt="Drawing" style="width: 500px;"/> </td>

**Layes (Tầng)**: Một MLP có các layers là Input layers, Output layers và có thể có nhiều hơn một Hidden layers. Số lượng layers của một MLP được tính bằng số layers trong mô hình - 1, tức không tính Input layers. Hình ví dụ có thể nhận thấy số layers $L=2$. 

**Units**: Một node (nút) trong một layer được gọi là một unit. Unit ở các input layer, hidden layers, và output layer được lần lượt gọi là input unit, hidden unit, và output unit. Đầu vào của các hidden layer được ký hiệu bởi 
$z$  đầu ra của mỗi unit thường được ký hiệu là $a$ (thể hiện activation, tức giá trị của mỗi unit sau khi ta áp dụng activation function lên $z$). Đầu ra của unit thứ $i$ trong layer thứ $l$ được kí hiệu là $a_i^{(l)}$. Giả sử thêm rằng số unit trong layer thứ l (không tính bias là $d^{(l)}$. Vector biểu diễn output của layer thứ $l$ được kí hiệu là $a^{(l)} \in R^{d^{(l)}}$

#### Activation Funtions 

Mỗi output của một unit (trừ các input units) được tính dựa vào công thức:
$$a_i^{(l)} = f(w_i^{(l)T}a^{(l-1)}+b_i^{(l)})$$
Trong đó $f(.)$ là một hàm activation function (nonlinear). Ở dạng vector, biểu thức trên được viết lại là: 
$$a^{(l)} = f(W^{(l)T}a^{(l-1)}+b^{(l)})$$
Khi activation function $f(.)$ được áp dụng cho một ma trận (hoặc một vector), ta hiểu rằng nó được áp dụng cho từng thành phần của ma trận đó. Sau đó các thành phần này được sắp xếp lại đúng theo thứ tự để được một ma trận có kích thước bằng với ma trận input.

Hàm sgn (còn gọi là hard-threshold) chỉ được sử dụng trong PLA, mang mục đích giáo dục nhiều hơn. Trong thực tế, hàm sgn không được sử dụng vì hai lý do: đầu ra là discrete, và đạo hàm tại hầu hết các điểm bằng 0 (trừ điểm 0 không có đạo hàm). Việc đạo hàm bằng 0 này khiến cho các thuật toán gradient-based (ví dụ như Gradient Descent) không hoạt động!

Hàm sigmoid có dạng $f(s) = 1/(1 + exp(-s))$ với đồ thị như trong Hình 5 (trái). Nếu đầu vào lớn, hàm số sẽ cho đầu ra gần với 1. Với đầu vào nhỏ (rất âm), hàm số sẽ cho đầu ra gần với 0. Hàm số này được sử dụng nhiều trong quá khứ ví có đạo hàm rất đẹp. Những năm gần đây, hàm số này ít khi được sử dụng. Nó có một nhược điểm cơ bản:

* Sigmoid saturate and kill gradients: Một nhược điểm dễ nhận thấy là khi đầu vào có trị tuyệt đối lớn (rất âm hoặc rất dương), gradient của hàm số này sẽ rất gần với 0. Điều này đồng nghĩa với việc các hệ số tương ứng với unit đang xét sẽ gần như không được cập nhật

Hàm tanh cũng có nhược điểm tương tự về việc gradient rất nhỏ với các đầu vào có trị tuyệt đối lớn.

Hình 5: Hàm ReLU và tốc độ hội tụ khi so sánh với hàm tanh.


ReLU (Rectified Linear Unit) được sử dụng rộng rãi gần đây vì tính đơn giản của nó. Đồ thị của hàm ReLU được minh họa trên Hình 5 (trái)). Nó có công thức toán học $f(s) = max(0,s)$ 
* ReLU được chứng minh giúp cho việc training các Deep Networks nhanh hơn rất nhiều (theo Krizhevsky et al.). Hình 5 (phải) so sánh sự hội tụ của SGD khi sử dụng hai activation function khác nhau: ReLU và tanh. Sự tăng tốc này được cho là vì ReLU được tính toán gần như tức thời và gradient của nó cũng được tính cực nhanh với gradient bằng 1 nếu đầu vào lớn hơn 0, bằng 0 nếu đầu vào nhỏ hơn 0.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
x = np.linspace(-1, 1, 200)[:, None]       # [batch, 1]
y = x ** 2                                  # [batch, 1]
learning_rate = 0.001


def tanh(x):
    return np.tanh(x)


def derivative_tanh(x):
    return 1 - tanh(x)**2


w1 = np.random.uniform(0, 1, (1, 10)) # 10
w2 = np.random.uniform(0, 1, (10, 10)) # 100
w3 = np.random.uniform(0, 1, (10, 1)) # 10


b1 = np.full((1, 10), 0.1)
b2 = np.full((1, 10), 0.1)
b3 = np.full((1, 1), 0.1)


for i in range(300):
    a1 = x
    z2 = a1.dot(w1) + b1 # (200,1)(1,10) = (200, 10)
    a2 = tanh(z2)
    z3 = a2.dot(w2) + b2 # (200, 10)(10, 10) = (200, 10)
    a3 = tanh(z3)
    z4 = a3.dot(w3) + b3 # (200, 10)(10, 1) = (200, 1)

    cost = np.sum((z4 - y)**2)/2

    # backpropagation
    z4_delta = z4 - y
    dw3 = a3.T.dot(z4_delta)
    db3 = np.sum(z4_delta, axis=0, keepdims=True)

    z3_delta = z4_delta.dot(w3.T) * derivative_tanh(z3)
    dw2 = a2.T.dot(z3_delta)
    db2 = np.sum(z3_delta, axis=0, keepdims=True)

    z2_delta = z3_delta.dot(w2.T) * derivative_tanh(z2)
    dw1 = x.T.dot(z2_delta)
    db1 = np.sum(z2_delta, axis=0, keepdims=True)

    # update parameters
    for param, gradient in zip([w1, w2, w3, b1, b2, b3], [dw1, dw2, dw3, db1, db2, db3]):
        param -= learning_rate * gradient
        
        
plt.plot(x,y)
plt.plot(x, z4)
plt.show()
```
