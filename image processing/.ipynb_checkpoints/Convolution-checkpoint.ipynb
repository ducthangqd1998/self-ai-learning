
**Convolution (Tích chập)**: là một khái niệm trong xử lý tín hiệu số nhằm biến đổi thông tin đầu vào thông qua một phép tích chập với bộ lọc để trả về đầu ra là một tín hiệu mới. Tín hiệu này sẽ làm giảm những đặc trưng mà bộ lọc không quan tâm và chỉ giữ những đặc trưng chính.

**Tích chập 2 chiều**: Tích chập thông dụng nhất là tích chập 2 chiều được áp dụng trên ma trận đầu vào và ma trận bộ lọc 2 chiều. Phép tích chập của một ma trận $X\in R^{W_1\times H_1}$ với một bộ lọc (receptive field) $F\in R^{F\times F}$ là một ma trận  $Y\in R^{W_2\times H_2}$ sẽ trả qua những bước sau:

**Tính tích chập tại 1 điểm**: Tại vị trí đầu tiên trên cùng của ma trận đầu vào ta sẽ lọc ra một ma trận con  $X\in R^{F\times F}$  có kích thước bằng với kích thước của bộ lọc. Giá trị  $y_{11}$  tương ứng trên $Y$ là tích chập của  $X$ với  $F$ được tính như sau:

$$y_{11} = \sum_{i=1}^F \sum_{j=1}^F x_{ij}f_{ij}$$

Tiến hành trượt dọc theo ma trận theo chiều từ trái qua phải, từ trên xuống dưới theo bước nhảy (stride) S ta sẽ tính được các giá trị  $y_{ij}$ tiếp theo. Sau khi quá trình này kết thúc ta thu được trọn vẹn ma trận  $Y$ .



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('images/btl.jpg', 0)
plt.axis('off')
plt.imshow(img, cmap='gray')
```

```python
def conv2d(img, kernel):
    w, h = img.shape[:2]
    wk, hk = kernel.shape[:2]
    ero = wk // 2 + 1
    conv = np.zeros((w-ero, h-ero))

    for i in np.arange(conv.shape[0]):
        for j in np.arange(conv.shape[1]):
            conv[i,j] = np.sum(img[i:i+wk, j:j+hk] * kernel) # Co the su dung ham tensordot
    
    return conv    
```

```python
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])

plt.axis('off')
plt.imshow(conv2d(img, kernel))
```

```python
# Kernel theo chieu doc
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])
filtered = cv2.filter2D(src=img, kernel=kernel, ddepth=-1)
plt.axis('off')
plt.imshow(filtered, cmap='gray')
```

```python
# Kernel theo chieu ngang
kernel = np.array([[1,  1,  1],
                   [0,  0,  0],
                   [-1, -1, -1]])
filtered = cv2.filter2D(src=img, kernel=kernel, ddepth=-1)
plt.axis('off')
plt.imshow(filtered, cmap='gray')
```

```python
# blur image
kernel = np.array([[1, 1, 1],
                   [1, 1, 1],
                   [1, 1, 1]]) / 9
filtered = cv2.filter2D(src=img, kernel=kernel, ddepth=-1)
plt.axis('off')
plt.imshow(filtered, cmap='gray')
```

```python
# sharpening image
# Cái này khá khó hiểu, kiểu trong 1 kernel, pixel ở giữa có trọng số cao hơn các pixel liền kề, 
# giúp nổi bật hơn các pixel đó, thành ra khi nhìn vào sẽ giống như bị gãy và ảnh nhìn sống động hơn.
kernel = (np.array([[-1, -1, -1],
                    [-1,  9, -1],
                    [-1, -1, -1]]))
filtered = cv2.filter2D(src=img, kernel=kernel, ddepth=-1)
plt.axis('off')
plt.imshow(filtered, cmap='gray')
```
