## Data Augmentation

Trong những năm gần đây, AI phát triển mạnh và cho các kết quả đầy khả quan. Một trong những lý do đó là sự ra đời của học sâu (deep learning), học sâu là bước đột phát của AI trong thời điểm hiện tại, đã hồi sinh lĩnh vực AI sau hàng chục năm lận đận. Nhưng cũng có rất nhiều thách thức cho học sâu, trong đó là vấn đề về dữ liệu. Rõ ràng đa số các mô hình hiện nay đều là các mô hình học giám sát và học bán giám sát, các mô hình này đòi hỏi một lượng dữ liệu lớn để huấn luyện. Nó quan trọng đến nỗi mà ta có thể nói độ lớn của dữ liệu tỉ lệ thuận với độ chính xác của mô hình. Việc thu thập dữ liệu thường rất tốn kém chi phí, do đó cần có một phương pháp có thể tăng số lượng lẫn sự phong phú của dữ liệu, từ đó **Data Augmentation** ra đời như là một phương pháp đơn giản và dễ dàng để giải quyết vấn đề trên.

Chúng ta sẽ tìm hiểu một vài kĩ thuật data augemtation phổ biến.

```python
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
import random
```

### Rotation

**Rotation**: Là kỹ thuật xoay ảnh, ảnh có thể xoay theo góc nghiêng. 

```python
def show_image(images, label):
    img_len = len(images)
    plt.figure(figsize=(10,10))

    plt.subplot(5, 3, 2)
    plt.title("Origin Image")
    plt.axis('off')
    img = cv2.cvtColor(images[0], cv2.COLOR_BGR2RGB)
    plt.imshow(img)

    for i in range(1, img_len):
        if i > 3:
            plt.subplot(5, 3, i + 9) 
        else:
            plt.subplot(5, 3, i + 6) 
        plt.title(label[i])
        plt.axis('off')
        img = cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.show()
```

```python
def rotation(image, angle=30):
    (h, w) = image.shape[:2]
    center = (w/2, h/2)
    M = cv2.getRotationMatrix2D(center, angle, scale = 1.0)
    return cv2.warpAffine(image, M, (w, h))
```

```python
img = cv2.imread('images/btl.jpg')
arr = []
title = []
for i in range(0, 7):
    rotate_img = rotation(img, i * 30)
    title.append('_'.join(["Rotate", str(i * 30)]))
    arr.append(rotate_img)
show_image(arr, title)
```

### Shear

**Shearing**: Là kỹ thuật thay đổi kích thước của bức ảnh. Nó khác với kỹ thuật Scaling ảnh thông thường là sẽ thay đổi hình dạng của ảnh thành đường chéo.

```python
def Shearing(image, angle):
    srcTri = np.array( [[0, 0], 
                        [image.shape[1] - 1, 0],
                        [0, image.shape[0] - 1]] ).astype(np.float32)
    dstTri = np.array( [[0, 
                         image.shape[1]*0.33], 
                        [image.shape[1]*0.85,
                         image.shape[0]*0.25], 
                        [image.shape[1]*0.15, 
                         image.shape[0]*0.7]] ).astype(np.float32)

    warp_mat = cv2.getAffineTransform(srcTri, dstTri)
    warp_dst = cv2.warpAffine(image, 
                              warp_mat, 
                              (image.shape[1], image.shape[0]))

    center = (warp_dst.shape[1]//2, warp_dst.shape[0]//2)
    scale = 1
    rot_mat = cv2.getRotationMatrix2D( center, angle, scale )
    return cv2.warpAffine(warp_dst, 
                          rot_mat, 
                          (image.shape[1], image.shape[0]))
```

```python
scale = random.randint(2, 7)
shear_img = Shearing(img, 30)
plt.axis('off')
shear_img = shear_img[:, :, ::-1]
plt.imshow(shear_img)
```

### Crop 
**Cropping**: Là kỹ thuật cắt ảnh, thường sẽ cắt ngẫu nhiên. Một điều lưu ý khi sử dụng kĩ thuật này là phải giữ được đối tượng trong ảnh. Nếu cắt ảnh mà không có đối tượng, xem như ảnh gắn nhãn sai

```python
def cropping(image):
    h, w = image.shape[:2]
    h_crop, w_crop = int(h * 0.6), int(w * 0.6)

    x = random.randint(0, h - h_crop)
    y = random.randint(0, w - w_crop)
    return image[y:y+h_crop, x:x+w_crop]
```

```python
plt.axis('off')
plt.imshow(cropping(img)[:, :, ::-1])
```

### Flip 
**Flipping**: Là kỹ thuật lật ảnh, ảnh có thể được lật theo chiều ngang hoặc chiều dọc  

```python
def flipping(image, direct=0):
    return cv2.flip(image, direct)
```

```python
vertical_image = flipping(img, 1)
horizontal_image = flipping(img, 0)
plt.axis('off')
plt.imshow(vertical_image[:,:,::-1])
```

```python
plt.axis('off')
plt.imshow(horizontal_image[:,:,::-1])
```

### Changing the brightness

```python
def Change_brightness(image, alpha, beta):
    new_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return new_image
```

```python
def gammaCorrection(image, gamma):
    lookUpTable = np.empty((1,256), np.uint8)
    for i in range(256):
        lookUpTable[0,i] = np.clip(pow(i / 255.0, gamma) * 255.0, 0, 255)

    res = cv2.LUT(image, lookUpTable)
    return res
```

```python
brightness_img = Change_brightness(img, 1.5, 20)
brightness_img_1 = gammaCorrection(img, 5)
plt.axis('off')
plt.imshow(brightness_img[:,:,::-1])
```
