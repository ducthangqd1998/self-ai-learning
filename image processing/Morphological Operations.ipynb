**Thresholding** là phương pháp giúp chuyển ảnh xám về ảnh nhị phân (black-white). Đây là một phương pháp đơn giản dùng để segment ảnh. 

````python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
````

## Thresholding

````python
img = cv2.imread('images/google.png', 0)
plt.imshow(img)
````

Sử dụng hàm: ```cv2.threshold(source, thresholdValue, maxVal, thresholdingTechnique)```
Trong đó:
* **source**: là ảnh, yêu cầu là ảnh xám 
* **thresholdValue**: Giá trị ngưỡng để thay đổi giá trị ảnh xám để quy về 255 hoặc là 0 
* **maxVal**: Giá trị lớn nhất có thể được gán cho 1 pixel
* **thresholdingTechnique**: Loại threshold được sử dụng

Một số  Thresholding Techniques:

* **cv2.THRESH_BINARY**: Giá trị pixel lớn hơn **thresholdValue** sẽ là 255, còn lại là 0.
* **cv2.THRESH_BINARY_INV**: Ngược lại với **cv2.THRESH_BINARY**.
* **cv2.THRESH_TRUNC**: Các giá trị lớn hơn hoặc bằng **thresholdValye** sẽ bằng **thresholding**, còn lại giữ nguyên.
* **cv2.THRESH_TOZERO**: Các pixel có giá trị nhỏ hơn **thresholding** sẽ bằng 0, còn lại giữ nguyeen.
* **cv2.THRESH_TOZERO_INV**: Ngược lại với **cv2.THRESH_TOZERO**.

````python
_, bi = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY) 
_, bi_inv = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY_INV) 
_, trunc = cv2.threshold(img, 80, 255, cv2.THRESH_TRUNC) 
_, zero = cv2.threshold(img, 80, 255, cv2.THRESH_TOZERO) 
_, zero_inv = cv2.threshold(img, 128, 255, cv2.THRESH_TOZERO_INV) 
````

````python
plt.axis('off')
plt.imshow(bi, cmap='gray')
````

````python
plt.axis('off')
plt.imshow(bi_inv, cmap='gray')
````

````python
plt.axis('off')
plt.imshow(trunc, cmap='gray')
````

````python
plt.axis('off')
plt.imshow(zero, cmap='gray')
````

````python
plt.axis('off')
plt.imshow(zero_inv, cmap='gray')
````

## Bitwise

Các phép tính bitwise được sử dụng rất nhiều trong xử lí hình ảnh, giúp lấy các thông tin cần thiết của ảnh/

````python
square_img = cv2.imread('images/square.png')
circle_img = cv2.imread('images/circle.png')
````

````python
plt.imshow(square_img)
````

````python
plt.axis('off')
plt.imshow(circle_img)
````

````python
dest_and = cv2.bitwise_and(square_img, circle_img, mask = None) 
plt.axis('off')
plt.imshow(dest_and)
````

````python
dest_and = cv2.bitwise_or(square_img, circle_img, mask = None) 
plt.axis('off')
plt.imshow(dest_and)
````

````python
dest_and = cv2.bitwise_xor(square_img, circle_img, mask = None) 
plt.axis('off')
plt.imshow(dest_and)
````

````python
dest_and = cv2.bitwise_not(circle_img, mask = None) 
plt.axis('off')
plt.imshow(dest_and)
````

## Erosion & Dilation

Erosion và Dilation là các phương pháp biến đổi dựa trên hình dạng, gọi chung là các phương pháp **Morphological**. Mục đích chủ yếu là làm giảm hoặc tăng sự nổi bật cho các đặc trưng của ảnh. 

````python
img = cv2.imread('images/google.png', 0)
plt.imshow(img)
````

````python
# Tạo một ma trận kernel có kích thước là 5x5 
kernel = np.ones((5,5), np.uint8)
````

````python
'''
cv2.erode và cv2.dilate đều có 3 tham số 
    + img: Ảnh nguồn
    + kernel: ma trận kernel 
    + interations: Số lần lặp
'''
img_erosion = cv2.erode(img, kernel, iterations=1) 
img_dilation = cv2.dilate(img, kernel, iterations=1) 
````

````python
plt.axis('off')
plt.imshow(img_dilation)
````

````python
plt.axis('off')
plt.imshow(img_erosion)
````
